```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
# data = pd.read_excel("code_card_data.xlsx")
data = pd.read_excel("count_card_data.xlsx")
data = data.drop_duplicates(subset=data.columns[:15].tolist())
data.head()
```

```python
data=data.to_numpy()
print(len(data))
```

```python
# X_train = data[0:1000, :54].astype(float)  # 前8行的前15列作为训练特征
# y_train = data[0:1000, 55].astype(float)   # 前8行的最后一列作为训练标签
X_train_, X_test_, y_train, y_test_true = train_test_split(data[0:10475, :16], data[0:10475, 16], test_size=0.1, random_state=1)
X_train = X_train_[:, :15]
X_test = X_test_[:, :15]
X_test_app = X_test_[:, 15]
# 训练随机森林分类器
clf = RandomForestClassifier(random_state=20)#n_estimators=25,n_jobs
clf.fit(X_train, y_train)
```

```python
# data1 = pd.read_excel("a113.xlsx")
# data1=data1.to_numpy()
```

```python
# X_test = data1[:, :15].astype(float)
# y_test_true = data1[:, 16].astype(float)
# # 预测测试数据
y_test_pred = clf.predict(X_test)

# 输出结果
# print("测试数据真实标签:", y_test_true)
# print("测试数据预测标签:", y_test_pred)
print("准确率:", accuracy_score(y_test_true, y_test_pred))
```

```python
k=0
for i in range(len(y_test_true)):
    if y_test_true[i]!=y_test_pred[i]:
        if abs(y_test_true[i]-y_test_pred[i])>1:
            k+=1
print(k/len(y_test_true))
```

```python
k1 = 0
for i in range(len(y_test_true)):
    if y_test_true[i]!=y_test_pred[i]:
        a = y_test_true[i] - X_test_app[i]
        b = y_test_pred[i] - X_test_app[i]
        if abs(a)>0.7 or abs(b)>0.7:
            k1+=1
print(1-k1/len(y_test_true))
```
