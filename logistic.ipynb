```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# data = pd.read_excel("count_card_data.xlsx")
data = pd.read_excel("code_card_data.xlsx")
data = data.drop_duplicates(subset=data.columns[:54].tolist())
```

```python
data=data.to_numpy()
```

```python
# X_train = data[0:3000, :15].astype(float)  
# y_train = data[0:3000, 16].astype(float)   
# # X_test = data[28:, :15].astype(float)  
# # y_test_true = data[28:, 15].astype(float)  

X_train_, X_test_, y_train, y_test_true = train_test_split(data[0:, :55], data[0:, 55], test_size=0.1, random_state=1)
X_train = X_train_[:, :54]
X_test = X_test_[:, :54]
X_test_app = X_test_[:, 54]
# 训练逻辑回归模型
logreg = LogisticRegression(random_state=42, max_iter=1000)
logreg.fit(X_train, y_train)
```

```python
# 预测测试数据
y_test_pred = logreg.predict(X_test)

# 输出结果
# print("测试数据真实标签:", y_test_true)
# print("测试数据预测标签:", y_test_pred)
print("准确率:", accuracy_score(y_test_true, y_test_pred))
```

```python
k1 = 0
for i in range(len(y_test_true)):
    if y_test_true[i]!=y_test_pred[i]:
        a = y_test_true[i] - X_test_app[i]
        b = y_test_pred[i] - X_test_app[i]
        if abs(a)>0.7 or abs(b)>0.7:
            k1+=1
print(1-k1/len(y_test_true))
```

```python
k=0
for i in range(len(y_test_true)):
    if y_test_true[i]!=y_test_pred[i]:
        if abs(y_test_true[i]-y_test_pred[i])>1:
            k+=1
print(k/len(y_test_true))
```

```python
# 查看模型系数（特征重要性）
print("\n特征系数（权重）:")
for i, coef in enumerate(logreg.coef_[0]):
    print(f"特征 {i+1}: {coef:.4f}")
```
