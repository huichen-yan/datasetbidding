```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# data = pd.read_excel("count_card_data.xlsx")
data = pd.read_excel("code_card_data.xlsx")
data = data.drop_duplicates(subset=data.columns[:54].tolist())
data.head()
```

```python
X_train, X_test, y_train, y_test_true = train_test_split(data[0:, :54], data[0:, 54], test_size=0.1, random_state=1)
# 创建并训练线性回归模型
model = LinearRegression()
model.fit(X_train, y_train)

# 预测测试集
y_pred = model.predict(X_test)

# 输出结果
# print("测试数据真实值:", y_test_true)
# print("测试数据预测值:", y_pred)
# print("均方误差(MSE):", mean_squared_error(y_pred, y_test_true))
```

```python
y = 0
for i in range(len(y_pred)):
    y += abs(y_test_true[i]-y_pred[i])
print(y/len(y_pred))
```
